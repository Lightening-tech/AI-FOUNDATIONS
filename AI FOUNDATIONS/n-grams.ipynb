```python
dataset = [
    "There's fire over the city."

    "Mongopark is in Africa."

    "The Nigerian dream is to leave."

]

def space_tokenize(text):
    return text.split()
all_unigrams = []
all_bigrams = []
all_trigrams = []

def generate_ngrams(text: str, n: int) -> list[tuple[str]]:
    """Generates n-grams from a given text.

    Args:
        text: The input text string.
        n: The size of the n-grams (e.g., 2 for bigrams, 3 for trigrams).

    Returns:
        A list of n-grams, each represented as a list of tokens.
    """

    # Tokenize text using the globally defined `space_tokenize` function.
    tokens = space_tokenize(text)

    # Construct the list of n-grams.
    ngrams = []

    # If the number of tokens is less than n, no n-grams can be formed.
    if len(tokens) < n:
        return ngrams

    # Iterate from the start of the token list up to the point
    # where a full n-gram can still be formed.
    for i in range(len(tokens) - n + 1):
      # Slice the token list from the current index i to i + n
      ngram_slice = tokens[i :i + n]
      # Convert the slice (list) to a tuple and add to our results
      ngrams.append(tuple(ngram_slice))

    return ngrams

for paragraph in dataset:
    # Calling `generate_ngrams` with n=1 constructs a list of unigrams.
    all_unigrams.extend(generate_ngrams(paragraph, n=1))
    # Calling `generate_ngrams` with n=2 constructs a list of bigrams (2-grams).
    all_bigrams.extend(generate_ngrams(paragraph, n=2))
    # Calling `generate_ngrams` with n=2 constructs a list of trigram (3-grams).
    all_trigrams.extend(generate_ngrams(paragraph, n=3))

print("First 10 Unigrams:", all_unigrams[:10])
print("First 10 Bigrams:", all_bigrams[:10])
print("First 10 Trigrams:", all_trigrams[:10])
```

```python
dataset = [
    "There's fire over the city, no one can sleep.",
    "Everyone around me said that nothing was heard or seen.",
    "Woke up in a city that couldn't sleep.",
    "It's burning everywhere and still nobody cares.",
    "You should see how the people believe still.",
    "They said help from Heaven will come soon.",
    "Oh oh brother, oh oh, I still remember, no good turn in the city.",
    "You no ready to be foolish, tell me how you wan take wise?",
    "Na the problem for this city, everybody dey claim wise.",
    "Look around you well well, nobody ready to fight the fight.",
]


def space_tokenize(text):
    return text.split()


all_unigrams = []
all_bigrams = []
all_trigrams = []


def generate_ngrams(text: str, n: int) -> list[tuple[str]]:
    """Generates n-grams from a given text.

    Args:
        text: The input text string.
        n: The size of the n-grams (e.g., 2 for bigrams, 3 for trigrams).

    Returns:
        A list of n-grams, each represented as tuples of tokens.
    """

    tokens = space_tokenize(text)
    ngrams = []

    if len(tokens) < n:
        return ngrams

    for i in range(len(tokens) - n + 1):
        ngram_slice = tokens[i : i + n]
        ngrams.append(tuple(ngram_slice))

    return ngrams


for paragraph in dataset:
    all_unigrams.extend(generate_ngrams(paragraph, n=1))
    all_bigrams.extend(generate_ngrams(paragraph, n=2))
    all_trigrams.extend(generate_ngrams(paragraph, n=3))


print("First 10 Unigrams:", all_unigrams[:10])
print("First 10 Bigrams:", all_bigrams[:10])
print("First 10 Trigrams:", all_trigrams[:10])
```
